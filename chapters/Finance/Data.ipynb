```python
import numpy as np
import pandas as pd
import wrds
import psycopg2 
from datetime import datetime
import pandas_datareader.data as web
conn=wrds.Connection()
```

# Warren Buffet and ARKK Data

```python


start = datetime(1960, 1, 1)  # Replace YYYY, MM, DD with the desired start date

ds = web.DataReader('F-F_Research_Data_5_Factors_2x3_Daily', 'famafrench', start=start)
dff5 = ds[0]

ds = web.DataReader('F-F_Momentum_Factor_Daily', 'famafrench', start=start)
dmom = ds[0]

dff6 = dff5.merge(dmom, left_index=True, right_index=True, how='inner')
dff6 = dff6 / 100
```

```python
from codesnippets import get_returns_daily
df=get_returns_daily(['BRK'],conn,'1/1/1960','1/1/2022',variable='ret')


df=(dff6).merge(df[17778.0],left_index=True,right_index=True,how='inner').dropna(how='any')
df.rename(columns={17778:'BRK'},inplace=True)


df_temp=get_returns_daily(['ARKK'],conn,'1/1/1960','1/1/2022',variable='ret')
df=(df).merge(df_temp,left_index=True,right_index=True,how='left')
df.rename(columns={14948.0:'ARKK'},inplace=True)
df.to_pickle('../../assets/data/df_WarrenBAndCathieW.pkl')
df
```

```python


start = datetime(1960, 1, 1)  # Replace YYYY, MM, DD with the desired start date

ds = web.DataReader('F-F_Research_Data_5_Factors_2x3', 'famafrench', start=start)
dff5 = ds[0]

ds = web.DataReader('F-F_Momentum_Factor', 'famafrench', start=start)
dmom = ds[0]

dff6 = dff5.merge(dmom, left_index=True, right_index=True, how='inner')
dff6 = dff6 / 100
dff6.index = dff6.index.to_timestamp() + pd.offsets.MonthEnd()
dff6

from codesnippets import get_returns_monthly
df=get_returns_monthly(['BRK'],conn,'1/1/1960','1/1/2022',variable='ret')


df=(dff6).merge(df[17778.0],left_index=True,right_index=True,how='inner').dropna(how='any')
df.rename(columns={17778:'BRK'},inplace=True)


df_temp=get_returns_monthly(['ARKK'],conn,'1/1/1960','1/1/2022',variable='ret')
df=(df).merge(df_temp,left_index=True,right_index=True,how='left')
df.rename(columns={14948.0:'ARKK'},inplace=True)
df.to_pickle('../../assets/data/df_WarrenBAndCathieW_monthly.pkl')
df
```
